```python
import sqlite3
```

```python
# connect sqlite3 database
conn = sqlite3.connect('ecommerce.db')
```

```python
# cursor
cursor = conn.cursor()
```

```python
try:
    cursor.execute(
        """
            Create table Category(
                Category_id integer primary key autoincrement,
                category_name text
            )
        """
    )
except:
    print('Error: Table Category Already Exists!!')
else:
    print('Table created successfully!!')
```

```python
# find details of the products whose price range is from 20000 to 50000.
cursor.execute(
    '''
       select * from product
       where product_price >= 20000 and product_price <= 50000
    '''
    # using between
    # select * from product
    # where product_price between 20000 and 50000;
)
data = cursor.fetchall()
for d in data:
    print(d)
```

```python
# find details of samsung and MI TV and smart watch 
# for all kinds of database numeric value -> AND , categorical value -> OR
cursor.execute(
    '''
       select * from product where
       product_name = 'samsung TV' or product_name = 'MI TV' or 
       product_name = 'smart watch';
    '''
    # using IN 
    # select * from product
    # where product_name in('samsung TV', 'MI TV', 'smart watch');
)
data = cursor.fetchall()
for d in data:
    print(d)
```

```python
#find details of products whose name must contain 7 letters and third letter should be 'm'
cursor.execute(
    '''
       select * from Product 
       where product_name like '__m____%';
    '''
)
data = cursor.fetchall()
for d in data:
    print(d)
```

```python
# find details of products whose name must start with letter 'S'.
cursor.execute(
    """
       select * from Product 
       where product_name like 'S%';

    """
)
data = cursor.fetchall()
for d in data:
    print(d)
```
