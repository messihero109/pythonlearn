ERROR HANDLING

```python
# try except else finally
try:
    a = 1/0
    print(a)
except:
    print("Number cannot be divisible")
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    num3 = num1/num2
    print(num3)
except ZeroDivisionError:
    print("Error: Number cannot be divisible by zero.")
except ValueError:
    print("Error: only integer(Number) is accepted in this program.")
```

CUSTOM EXCEPTION

```python
# check age
class AgeError(Exception): # convert class into exception class
    pass
```

```python
class AgeError(Exception):
    pass
def check_age(age):
    if age >= 18:
        print('You are eligible for voting.')
    else:
        raise AgeError(f'Provided age {age} is not eligible for voting.')
try:
    age = int(input("Enter your age: "))
    check_age(age)
except AgeError as e: # e is used to print specific comments raised 
    print(e)
```
