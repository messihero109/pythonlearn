```python
# arithmetic operations
a = 10
b = 20
c = a + b
d = b // a
e = b ** a
f = b % a # modulus operator to get remainder
print(d)
print(c)
print(e)
print(f)
```

```python
# Logical operators
# we use these operators to combine conditional statements
# boolean values -> True or False
num1 = 10
num2 = 20
num3 = 30
```

```python
# AND 
print(num1 < num2 and num2 < num3) # True and True -> True

```

```python
# OR 
print(num1 < num2 or num2 <num3) # True or False -> True
```

```python
# == -> equal to
# != -> not equal to
a = 10
b = 20 
print(a == b) 
print(a != b)
```
