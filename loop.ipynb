```python
# Program should run only 5 times
# secret_number = 7
# USer should see an input field to enter guess number
# if user guess is correct number then print -> Number Guessed Correctly
# if user guess wrong number then program should show input field again.
secret_number = 7
guess = int(input("Enter your guess number: "))
for i in range(5):
    if guess == secret_number:
        print("Number Guessed Correctly")
        break
    else:
        print("Input field again.")
```
