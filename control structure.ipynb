STUDENT GRADE RESULT

```python
marks = 30
if marks <= 80 and marks >= 100:
   print("Grade: A")
elif marks <= 70 and marks >= 80:
   print("Grade: B")
elif marks <= 60 and marks >= 70:
   print("Grade: C")
elif marks <= 40 and marks >= 60:
   print("Grade: D")
else:
    print("Fail")
```

Income Tax Calculator

```python
annual_income = float(input("Enter your annual income: "))
if annual_income <= 250000:
    print("No tax")
elif annual_income > 250001 and annual_income <= 500000:
    tax = annual_income * 0.05
    print("Taxable  income:", tax)
    print("Remaining Balance:", annual_income - tax)
elif annual_income > 500001 and annual_income <= 1000000:
    tax = annual_income * 0.20
    print("Taxable income:", tax)
    print("Remaining Balance:", annual_income - tax)
else:
    tax = annual_income * 0.30
    print("Taxable income:", tax)
    print("Remaining Balance:", annual_income - tax)
```

Exam Eligibilty system (Nested if) # Nested if is a condition in which : You put an if statement inside another if, elif, or else block. # One decision inside another decision.

```python
Attendance = float(input("Enter your attendance : "))
Internal_marks = float(input("Enter your internal marks: "))
if Attendance >= 75:
    if Internal_marks >= 40:
        print("You are eligible")
    else:
        print("You are not eligible for the exam due to low internal marks")
else:
    print("Not Eligible for the exam due to low attendance")
        
```

Temperature Measurement

```python
Temperature = float(input("Enter the temperature in Celsius: "))
if Temperature >= 40:
    print("It's hot today!")
elif Temperature >= 25 and Temperature <= 40:
    print("It's warm today!")
else:
    print("It's cold today!")
```

Bank Loan Approval System

```python
Age = int(input("Enter your age: "))
Monthly_income = float(input("Enter your monthly income: "))
if Age >= 18:
    if Monthly_income >= 30000:
        print("Loan approved")
    else:
        print("Loan rejected due to insufficient income")
else:
    print("Loan rejected due to age")
```

Withdraw Amount(ATM)

```python
balance = 50000
withdraw = int(input('Enter withdraw amount: '))
 
if withdraw < balance: # 500 < 50000 -> True
    if withdraw % 1000 == 0: # 500 % 1000 == 500
        print(f'Balance: {balance}') # f -> F-String
        balance = balance - withdraw
        print(f'Amount withdrawn: {withdraw}')
        print(f'Remaining Balance: {balance}')
    else:
        print('Withdraw amount must be multiple of 1000.')
else:
    print('Withdraw amount must be less than balance amount.')
```
